```python
import time
import os

# Ensure the directory exists
os.makedirs("homework5", exist_ok=True)

# File paths
hdf_file = "homework5/data.h5"
feather_file = "homework5/data.feather"
parquet_file = "homework5/data.parquet"
time_file = "homework5/times.txt"

# Dictionary to store times
times = {}

# Write to HDF
start = time.time()
df.to_hdf(hdf_file, key="data", mode="w")
times["HDF"] = time.time() - start

# Write to Feather
start = time.time()
df.to_feather(feather_file)
times["Feather"] = time.time() - start

# Write to Parquet
start = time.time()
df.to_parquet(parquet_file)
times["Parquet"] = time.time() - start

# Save times to a text file
with open(time_file, "w") as f:
    for format_name, duration in times.items():
        f.write(f"{format_name}: {duration:.6f} seconds\n")

# Print the times
print("Write times (seconds):")
for format_name, duration in times.items():
    print(f"{format_name}: {duration:.6f}")
```
